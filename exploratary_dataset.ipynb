```python
# Load pickled data
import pickle

training_file = 'traffic-signs-data/train.p'

with open(training_file, mode='rb') as f:
    train = pickle.load(f)
    
X_train, y_train = train['features'], train['labels']
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

objects = set(y_train)
y_pos = np.arange(len(objects))
no_of_records_per_sign = []

train = np.array(y_train)
#This would give array of unique counts for labels
no_of_records_per_sign = np.bincount(train)

df = pd.read_csv("signnames.csv", names = ["number", "name"], skiprows = 1)
print(df.head(5))

print()

object_names = []

for i in objects:
    object_names.append(df.loc[df["number"] == i]["name"].values[0])

plt.barh(y_pos, no_of_records_per_sign, align='center', alpha=0.5)
plt.yticks(y_pos, object_names, fontsize = 6)
plt.xlabel('No of records')
plt.title('Exploratory visualization')


plt.savefig('Exploratory_visualization.png')  
plt.show()  
plt.close()
```
